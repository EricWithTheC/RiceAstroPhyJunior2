```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20*np.pi, 1000)
y = np.sin(x)
yfinal = y + 0.1*np.random.normal(size=x.size)
def expSmooth(y, alpha):
    ysmooth = np.zeros(y.size)
    ysmooth[0] = y[0]
    for i in range(1, y.size):
        ysmooth[i] = alpha*y[i] + (1-alpha)*ysmooth[i-1]
    return ysmooth

plt.plot(x, yfinal, label='Noisy')
plt.plot(x, expSmooth(yfinal, 0.1), label='alpha=0.1')
```
