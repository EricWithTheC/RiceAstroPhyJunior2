```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def enl(n, l):
    R = 1
    k = 0.01
    q = 1
    return R/n**2 + (k*q/(4*R))*(3*n**2-l*(l+1))
```

```python
for i in [1,2,3]:
    for j in range(i):
        n = i
        l = j
        Enl = enl(n, l)
        plt.plot(n, Enl, 'ro')
```
