```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rydberg = 13.605693009*1.60217662e-19
a0 = 0.52917721067e-10 # Bohr radius in meters
e = 1.60217662e-19 # electron charge in Coulombs
hbar = 1.0545718e-34 # Reduced Planck constant in J s
E0 = 1
m = 9.10938356e-31 # electron mass in kg

def TTR(k):
    return k**3/(1+k**2)**6
```

```python
klist = np.arange(0, 10, 0.1)
TTRlist = TTR(klist)
plt.plot(klist, TTRlist)
plt.xlabel('k')
plt.ylabel('TTR')
plt.title('Total Transmission Rate')
# plt.yscale('log')
plt.show()
```

```python
f = (hbar/(6* m* a0**2) + rydberg/ hbar)/(2* np.pi)
```
