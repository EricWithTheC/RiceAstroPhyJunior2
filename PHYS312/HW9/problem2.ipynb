```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def P(Omega, omega):
    return ((Omega**4)/(16*omega**2*(omega**2-Omega**2)**2))*(np.sin(np.pi*omega/Omega)**2)
```

```python
omega = 1
Omegas = np.arange(0,5,0.01)
plt.figure(figsize = (8,6))
plt.title(r'Scaled Transition Probability ($0 < \Omega < 5\omega$)', fontsize = 14)
plt.plot(Omegas, P(Omegas, omega))
plt.xlabel(r'$\Omega / \omega$', fontsize = 14)
plt.ylabel(r'$P$', fontsize = 14)
plt.show()
```

```python
omega = 1
Omegas = np.arange(0,0.5,0.001)
plt.figure(figsize = (8,6))
plt.title(r'Scaled Transition Probability ($0 < \Omega < \frac{1}{2}\omega$)', fontsize = 14)
plt.plot(Omegas, P(Omegas, omega))
plt.xlabel(r'$\Omega / \omega$', fontsize = 14)
plt.ylabel(r'$P$', fontsize = 14)
plt.show()
```
