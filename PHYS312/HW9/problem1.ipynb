```python
plt.figure(figsize=(5, 50))
plt.title('Vibrational and Rotational Energy Levels')
for i in energyLevels:
    plt.axhline(y=i, color='r', linestyle='-')

plt.xlabel('?')
plt.ylabel('Energy')

```
